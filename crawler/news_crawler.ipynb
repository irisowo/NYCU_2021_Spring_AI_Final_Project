```python
import csv
import firebase_admin
from firebase_admin import credentials, firestore

from datetime import datetime, timezone, timedelta
import datetime
from opencc import OpenCC
cc = OpenCC('t2s')

cred = credentials.Certificate(path + "test-2a741-firebase-adminsdk-qusio-846a25578a.json")
default_app = firebase_admin.initialize_app(cred)
media_list = ["chinatimes", "cna", "cts", "ebc", "ettoday", "ltn", "setn", "storm", "udn"]
keys = ["title", "media", "category", "content", "pubdate", "url"]
keys_index = ["index", "title", "media", "category", "content", "pubdate", "url"]
label_keys = ["title", "media", "category", "content", "label"]
label_first = ["index", "title", "media", "category", "content", "label"]
db = firestore.client()

label_type = "2"
out_filename = "news_all_sim.csv"

if label_type == "1":
  out_filename = "news_label_pos.csv"
elif label_type == "0":
  out_filename = "news_label_neu.csv"
elif label_type == "-1":
  out_filename = "news_label_neg.csv" 
# with open(out_filename, 'a', newline='', encoding='utf-8-sig') as csvfile:
#     writer = csv.writer(csvfile)
#     writer.writerow(label_first)

with open(out_filename, 'a', newline='', encoding='utf-8-sig') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(keys_index)
count = 0
```

```python
# for media in media_list:

cities_ref = db.collection(u'news')
query = cities_ref.order_by(
    u'pubdate', direction=firestore.Query.DESCENDING).limit(20000)
docs = query.stream()
# docs = db.collection(u'news').stream()
# print(media)
for i, doc in enumerate(docs):
    # if(count>=200):
    #   break
    tmp = doc.to_dict()
    news_list = [count]
    count+=1
    # for key in keys:
    for key in keys:
        if key in tmp:
          if key =="title":
            news_list.append(cc.convert(tmp[key]))
          elif key =="content":
            b = ""
            for a in tmp[key]:
              b+=a
            news_list.append(b)
          else:
            news_list.append(tmp[key])
        else:
            news_list.append("NA")
    with open(out_filename, 'a', newline='', encoding='utf-8-sig') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(news_list)
```
